```python
# %reset -f
# import pprint
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain.schema import AIMessage, HumanMessage, SystemMessage

load_dotenv()
```

```python
from langchain_upstage import ChatUpstage
model = ChatUpstage()
```

```python
chat_history = []  # 메세지 (대화기록) 저장할 리스트 

# System message 설정 (옵션)
system_message = SystemMessage(content="You are a helpful AI assistant.")
chat_history.append(system_message)  # system message 추가
```

```python
# Chat loop
while True:
    # 사용자 질문 입력
    query = input("You: ")
    if query.lower() == "exit":
        break
    chat_history.append(HumanMessage(content=query))  # 대화 기록(prompt) 추가
    print(f"User: {query}")
    
    # Ai 답변 생성
    result = model.invoke(chat_history) # 대화 기록 참고해서 invoke
    response = result.content
    print(f"AI: {response}")
    chat_history.append(AIMessage(content=response))  # 답변 기록 추가
   
```

```python
# Print chat history
print("---- Message History ----")
print(chat_history)
```

#### ✅ 대화 요약하거나 최근 몇개의 메세지만 유지하자 

```python
# 메세지 (대화기록) 저장할 리스트
chat_history = []

MAX_HISTORY_LENGTH = 5 # 최근 대화기록 3개정도만 유지

# Chat loop
while True:
    # 사용자 질문 입력
    query = input("You: ")
    if query.lower() == "exit":
        break
    chat_history.append(HumanMessage(content=query))  # 대화 기록(prompt) 추가
    print(f"User: {query}")
    
    # 최근 MAX_HISTORY_LENGTH 개의 메시지만 유지
    if len(chat_history) > MAX_HISTORY_LENGTH:
        chat_history = chat_history[-MAX_HISTORY_LENGTH:] #마지막 요소부터 (n개)만 선택
    
    # AI 답변 생성
    result = model.invoke(chat_history)  # 대화 기록을 모델에 전달하여 응답 생성
    response = result.content
    print(f"AI: {response}")
    chat_history.append(AIMessage(content=response))  # 답변 기록 추가
```
